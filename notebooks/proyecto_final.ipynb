PARA REALIZAR UN MODELO DE MACHINE LEARNING CAMBIAREMOS EL ENFOQUE DEL DATASET AGREGANDOLE UNA VARIABLE QUE SEA EL RATIO DE LA MEDIA DE INGRESOS Y EL PRECIO DE LOS VEHICULOS. 

```python
income_ref = 34716

prop_precio_ingreso = df_master["price_eur"] / income_ref
```

```python
df_master["categoria"] = prop_precio_ingreso.apply(fs.clasificar_accesibilidad)
```

FILTRAMOS EL DF PARA QUE SOLO QUEDEN LOS COCHES FAMILIARES.

```python
df_ml = df_master[df_master["coche_familiar"] == 1].copy()
```

VERIFICAMOS DISTRIBUCION

```python
df_ml["categoria"].value_counts(normalize=True)
```

AVANZAMOS EN EL PROYECTO PREPARANDO UN MODELO DE CLASIFICACION EL CUAL NOS PERMITA VALORAR LOS COCHES QUE SON ACCESIBLES PARA UNA FAMILIA CON INGRESO PROMEDIO.

```python
df_ml.head()
```

DEFINIMOS LAS VARIABLES QUE USARA EL MODELO, NUESTRA VARIABLE TARGET SERA "categoria"

```python
features_ml = ["price_eur", "costo_total_anual", "boot_capacity_l", "seating_capacity", "safety_rating", "sustainable", "powertrain_type"]
```

```python
features_ml = df_ml[features_ml]
target_ml = df_ml["categoria"]
```

```python
features_ml = pd.get_dummies(features_ml, columns=["powertrain_type"], drop_first=True)
```

ENTRENAREMOS DOS MODELOS: LOGISTIC REGRESSION Y RANDOM FOREST. 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features_ml, target_ml, test_size=0.2, random_state=42, stratify=target_ml)
```

```python
from sklearn.linear_model import LogisticRegression

model_ml = LogisticRegression(
    max_iter=2000,
    class_weight="balanced"
)

model_ml.fit(X_train, y_train)
```

```python
y_pred = model_ml.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

LOS RESULTADOS ARROJAN UNA ACCURACY DEL 91%, LO QUE INDICA QUE LOGRA CLASIFICAR CORRECTAMENTE EL MODELO. A SU VEZ EL RECALL DE LOS COCHES ACCESIBLES ES DEL 85% LO NOS MUESTRA QUE TIENE UNA DETECCIÓN ALTA Y LA PRECISION QUE EL 77% DE LAS VECES RECONOCE EL COCHE DE FORMA CORRECTA.

PROCEDEREMOS A GUARDAR EL MODELO PARA SUBIRLO AL STREAMLIT Y QUE EL MODELO SEA INTERACTIVO.

```python
import joblib

joblib.dump(model_ml, "../app/modelo_accesibilidad.pkl")
joblib.dump(features_ml.columns.tolist(), "../app/model_features.pkl")
```
